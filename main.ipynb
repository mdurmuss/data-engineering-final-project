## Sütun İsimleri Düzenleme

```python
df.rename(columns={'vSegID': 'SegmentID',
                   'vSegDir': 'SegmentDirection',
                   'fusedYear': 'Year',
                   'fusedMonth': 'Month',
                   'fusedday' : 'Day',
                   'fusedHour' : 'Hour' ,
                   'avgspeed': 'AverageSpeed',
                   'GRP' : 'Minute'
                  },
          inplace=True)
```

#### Aralıkları 15.dakika veya 45.dakika olarak işaretleyelim.

```python
GRPdict = {"0-30DK": 15 , "30-60DK": 45}
df['Minute'] = df['Minute'].apply(lambda x : GRPdict[x])

df.head()
```

### Yıl, Ay, Gün, Saat ile DateTime Bilgisi Oluşturma

```python
df['DateTime'] = pd.to_datetime(df[['Year','Month','Day','Hour','Minute']])
df.head()
```

## Yıllık Tatil Günleri Sütunu

```python
holidays = pd.read_excel("resmi_tatiller.xlsx")
holiday_dates = list(holidays['tarih'])

holiday_dates
```

```python
df['IsHoliday'] = df['DateTime'].apply(lambda x : True if str(x.date()) in holiday_dates else False)
```

```python
df.tail(10)
```

### Datetime Kullanarak Haftasonu Sütunu Ekleme

```python
df['DayName'] = df.apply(lambda tr: tr['DateTime'].day_name(), axis=1) 
```

```python
df['IsWeekend'] = df.apply(lambda tr: tr['DayName'] in ["Saturday", "Sunday"], axis=1)
```

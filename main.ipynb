```python
import pandas as pd
import numpy as np
```

## Kolon Adları
- vSegID	Segment bölgesini belirtir
- vSegDir	YÖN: 0:Geliş; 1:Gidiş
- fusedYear	Yıl
- fusedMonth	Ay
- fusedday	Gün
- fusedHour	Saat
- avgspeed	km'saat
- GRP	30'dklık dilimler
- shape	Segment ıdların koordinat bilgileri

### NOT: Elimizde hem Beşiktaş'a özel bir veri(besiktasHız.csv) hem de tüm ilçelerin karışık olduğu bir veri (fuseedİlce.csv) var. Bu sebeple ikisini birleştiriyoruz.

```python
BESIKTAS_DATA = "./BesiktasHiz.csv"
FUSEED_DATA = "./Fuseed_Data_İlçe.csv"
```

```python
besiktas_df = pd.read_csv(BESIKTAS_DATA, sep=";")
```

```python
besiktas_df.head()
```

```python
fuseed_df = pd.read_csv(FUSEED_DATA)
```

```python
fuseed_df.head()
```

```python
fuseed_besiktas_df = fuseed_df[fuseed_df.vSegID.isin(besiktas_df.vSegID)]
```

```python
fuseed_besiktas_df.head()
```

```python
df = besiktas_df.append(fuseed_besiktas_df)
```

```python
df.head()
```

## Toplam Satır Sayıları

```python
print(f'Besiktas veriseti: {len(besiktas_df.index)} satır')
print(f'Fuseed verisetindeki Besiktas verisi: {len(fuseed_besiktas_df.index)} satır')
print(f'Toplam Besiktas: {len(df.index)} satır')
```
